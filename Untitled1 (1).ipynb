```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\sudoko.jpg")
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges=cv2.Canny(gray,50,150)
lines=cv2.HoughLines(edges,1,np.pi/180,120)
for line in lines:
    rho,theta=line[0]
    a=np.cos(theta)
    b=np.sin(theta)
    x0=a*rho
    y0=b*rho
    x1=int(x0+1000*(-b))
    y1=int(y0+1000*(a))
    x2=int(x0-1000*(-b))
    y2=int(y0-1000*(a))
#     print("rho",rho,"theta",theta)
#     print("a=",a,"b=",b)
#     print(x0,y0)
#     print(x1,y1)
#     print(x2,y2)
    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
plt.imshow(img[...,::-1])
```

```python
original_img=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\0013.jpeg",cv2.IMREAD_COLOR)
img=original_img.copy()
blured=cv2.GaussianBlur(original_img,(5,5),0)
gray=cv2.cvtColor(blured,cv2.COLOR_BGR2GRAY)
edges=cv2.Canny(gray,50,150)
lines=cv2.HoughLinesP(edges,1,np.pi/180,50,minLineLength=20,maxLineGap=50)
for line in lines:
    x1,y1,x2,y2=line[0]
    cv2.line(img,(x1,y1),(x2,y2),(255,0,0),2)
plt.figure(figsize=[17,8])
plt.subplot(131);plt.imshow(original_img[...,::-1]);plt.title("original_img");
plt.subplot(132);plt.imshow(edges,cmap="gray");plt.title("original_img");
plt.subplot(133);plt.imshow(img,cmap="gray");plt.title("original_img");
```

```python
original_img=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\s.png",cv2.IMREAD_COLOR)
gray=cv2.cvtColor(original_img,cv2.COLOR_BGR2GRAY)
gray=cv2.medianBlur(gray,5)
rows=gray.shape[0]
circles=cv2.HoughCircles(gray,cv2.HOUGH_GRADIENT,1,rows/8,param1=200,param2=30)
if circles is not None:
    circles=np.uint16(np.around(circles))
    for c in circles[0,:]:
        center_x,center_y,radius=c
        center=(center_x,center_y)
        cv2.circle(original_img,center,radius,(255,0,255),3)
plt.imshow(original_img[...,::-1])
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img_fn=[r"C:\\Users\\Student\\Desktop\\pic\\a1.png",r"C:\\Users\\Student\\Desktop\\pic\\a2.png",
        r"C:\\Users\\Student\\Desktop\\pic\\a3.png",r"C:\\Users\\Student\\Desktop\\pic\\a4.png"]
img_list=[cv2.imread(fn)for fn in img_fn]
exposure_time=np.array([15.0,2.5,0.25,0.0333],dtype=np.float32)
plt.figure(figsize=[18,4])
for i ,(image,exposure) in enumerate(zip(img_list,exposure_time)):
    plt.subplot(1,4,i+1);plt.imshow(image[...,::-1]);plt.title(exposure);plt.xticks([]);plt.yticks([]);
plt.show()
```

```python
merge_mertens=cv2.createMergeMertens()
res_mertens=merge_mertens.process(img_list)
rs_mertens_8bit=np.clip(res_mertens*255,0,255).astype("uint8")
plt.figure(figsize=[17,8])
plt.imshow(rs_mertens_8bit[...,::-1]),plt.title("new")
```
