```python
def var(image):
    return cv2.Laplacian(image,cv2.CV_64F).var()
image1=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\004.jpeg",0)
image2=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\0014.jpg",0)
img1=var(image1)
img2=var(image2)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image1,cmap="gray");plt.title(img1)
plt.subplot(122);plt.imshow(image2,cmap="gray");plt.title(img2)
plt.show()
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    metric=var(gray)
    color=(0,0,255)
    if metric>200:
        color=(0,255,0)
    cv2.putText(frame,str(metric),(30,30),cv2.FONT_HERSHEY_COMPLEX,1,color)
    
    cv2.imshow("webcam",frame)
    if cv2.waitKey(1) &0xFF==27:
        break
cap.release()
cv2.destroyAllWindows()
```
